```python
from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM

# Load the pre-trained model and tokenizer from Hugging Face
tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-cnn")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/bart-large-cnn")

# Example text
text = '''
The transformers library provides general-purpose architectures for Natural Language Processing (NLP) tasks.
Initially, transformers like GPT, BERT, and T5 were built for specific NLP tasks such as language modeling,
classification, or translation. However, these architectures have shown excellent performance in text generation
tasks like summarization, which is the task of creating concise versions of longer documents.
Recent models like BART and PEGASUS are particularly suitable for abstractive text summarization.
'''

# Tokenize input text
inputs = tokenizer([text], max_length=1024, return_tensors="pt", truncation=True)

# Generate summary (using beam search for high-quality summary generation)
summary_ids = model.generate(inputs['input_ids'], num_beams=4, min_length=30, max_length=150, early_stopping=True)
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

print("Original Text:", text)
print("\nSummary:", summary)
```
